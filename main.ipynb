```python
from tensorflow.keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python

x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python

x_train = x_train.reshape(-1, 28*28)
x_test = x_test.reshape(-1, 28*28)
```

```python

from tensorflow.keras.utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model_relu = Sequential()

model_relu.add(Dense(128, activation='relu', input_shape=(784,)))
model_relu.add(Dropout(0.3))
model_relu.add(Dense(64, activation='relu'))
model_relu.add(Dropout(0.3))
model_relu.add(Dense(10, activation='softmax'))

model_relu.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

history_relu = model_relu.fit(
    x_train, y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)

loss_relu, acc_relu = model_relu.evaluate(x_test, y_test)

print("ReLU Results:")
print("Accuracy:", acc_relu)
print("Loss:", loss_relu)
```
